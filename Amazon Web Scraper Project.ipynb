```python
# getting asin from the source page: https://youtu.be/WIExbhe_GWc 

asin = []
items_list = []

driver = webdriver.Chrome(ChromeDriverManager().install())

for page_nb in range(1,3):
    driver.get('https://www.amazon.com/s?i=stripbooks&bbn=283155&rh=n%3A283155%2Cp_n_availability%3A2661600011&dc&page={}'.format(page_nb))
    
    # getting asins from the source page
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")
    products = soup.find_all('div', class_="s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col s-widget-spacing-small sg-col-12-of-16")
    for product in products:
        asin.append(product.attrs['data-asin'])
        
    # getting other features    
    tree = lxml.html.fromstring(driver.page_source)
    for product_tree in tree.xpath('//div[contains(@data-cel-widget, "search_result_")]'):
        title = product_tree.xpath('.//span[@class="a-size-medium a-color-base a-text-normal"]/text()')
            
        ratings = product_tree.xpath('.//span[@class="a-size-base s-underline-text"]/text()')
        reviews = product_tree.xpath('.//span[@class="a-icon-alt"]/text()')
        price = product_tree.xpath('.//span[@class="a-offscreen"]/text()')
            
        items = {
            'title': title,
            'ratings' : ratings,
            'reviews': reviews,
            'price' : price
        }
        items_list.append(items)
        
```
